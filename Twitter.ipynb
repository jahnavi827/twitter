```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.metrics import classification_report
```

```python
# Convert to PyTorch tensors
train_features = torch.tensor(X_train_scaled, dtype=torch.float)
test_features = torch.tensor(X_test_scaled, dtype=torch.float)
train_labels = torch.tensor(y_train, dtype=torch.long)
y_test_array = y_test.values
test_labels = torch.tensor(y_test_array, dtype=torch.long)
```

```python
# Edge index for a simple chain graph
edge_index = torch.tensor([[i, i + 1] for i in range(18)], dtype=torch.long).t().contiguous()
```

```python
# Mask creation for training and testing
num_nodes = train_features.size(0)  # Total number of nodes
train_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)
```

```python
# Randomly selecting 80% of the indices for training
perm = torch.randperm(num_nodes)
train_indices = perm[:int(0.8 * num_nodes)]
test_indices = perm[int(0.8 * num_nodes):]
```

```python
train_mask[train_indices] = True
test_mask[test_indices] = True
```

```python
# Creating the data objects
data = Data(x=train_features, edge_index=edge_index, y=train_labels, test_mask=test_mask)
data.train_mask = train_mask
data.test_mask = test_mask
```

```python
# GCN Model Definition
class GCN(torch.nn.Module):
    def __init__(self, num_features, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_features, 16)  # First GCN layer
        self.conv2 = GCNConv(16, num_classes)  # Second GCN layer to output class scores

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
# Initializing the model
model = GCN(num_features=12, num_classes=3)
class_counts = torch.bincount(data.y[data.train_mask])
class_weights = 1.0 / class_counts.float()  # Convert to float for precise division
class_weights /= class_weights.sum()  # Normalize the weights
```

```python
print(class_weights)
```

```python
# Using the calculated class weights in the loss function
criterion = torch.nn.NLLLoss(weight=class_weights)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
# Training Function
def train():
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Testing Function with Classification Report and Confusion Matrix
def test_and_report():
    model.eval()
    with torch.no_grad():
        out = model(data)
        pred = out.argmax(dim=1)
        correct = pred[data.test_mask].eq(data.y[data.test_mask]).sum().item()
        test_total = data.test_mask.sum().item()
        accuracy = correct / test_total

        # Generating classification report
        report = classification_report(data.y[data.test_mask].cpu().numpy(),
                                       pred[data.test_mask].cpu().numpy(),
                                       target_names=['Class 0', 'Class 1', 'Class 2'])

        # Generating confusion matrix
        cm = confusion_matrix(data.y[data.test_mask].cpu().numpy(),
                              pred[data.test_mask].cpu().numpy())
        return accuracy, report, cm

loss = train()
accuracy, report, cm = test_and_report()
print(f'Loss: {loss.item()}, Test Accuracy: {accuracy}')
print(report)
print("Confusion Matrix:")
sns.heatmap(cm_rf, annot=True, cmap='Blues', fmt='g', xticklabels=['Predicted Normal', 'Predicted Hate'],
            yticklabels=['True Normal', 'True Hate'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```
